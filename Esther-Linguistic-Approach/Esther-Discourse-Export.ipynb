```python
def findtexttype(sentencenode, data, pd):    
    # íŠ¹ì • ê°’ì´ ìˆëŠ” ì—´ì˜ ë‹¤ë¥¸ ê°’ì„ ì°¾ê¸°
    target_value = sentencenode  
    search_column = 'sentence_node'

    # ì¡°ê±´ì— ë§ëŠ” í–‰ ì„ íƒ
    search_results = data[data[search_column] == target_value]
    
    # ê²€ìƒ‰ ê²°ê³¼ ì¶œë ¥
    for index, row in search_results.iterrows():
        # í•´ë‹¹ í–‰ì˜ ë‹¤ë¥¸ ì—´ ê°’ ê°€ì ¸ì˜¤ê¸°
        value1 = row['verse']
        value2 = row['speakers']
        value3 = row['type']
        print(value1, value2, value3)
```
